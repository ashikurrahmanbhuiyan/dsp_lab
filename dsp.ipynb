```python
# convolution.py
import matplotlib.pyplot as plt

def convolution(x, h):
    N = len(x)
    M = len(h)
    output_size = N + M - 1
    y = [0] * output_size

    for n in range(output_size):
        for k in range(N):
            if 0 <= n - k < M:
                y[n] += x[k] * h[n - k]
    return y

# Define sequences
x = [1, 2, 3]  # Input sequence
h = [4, 5, 6]  # Impulse response

# Get convolution result
result = convolution(x, h)

# Plot the result
plt.figure(figsize=(10, 4))
plt.stem(result, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel("Index (n)")
plt.ylabel("Amplitude")
plt.title("Convolution of x and h")
plt.grid(True)
plt.show()

```

```python
#autocorrelation.py
import matplotlib.pyplot as plt

def autocorrelation(x):
    N = len(x)
    autocorr = [0] * N
    for m in range(N):
        autocorr[m] = sum(x[n] * x[n + m] for n in range(N - m))
    return autocorr

# Define sequence
x = [1, 2, 3, 4]  # Example sequence

# Compute autocorrelation
result = autocorrelation(x)

# Plot the result
plt.figure(figsize=(10, 4))
plt.stem(result, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel("Lag (m)")
plt.ylabel("Autocorrelation")
plt.title("Autocorrelation of the Sequence")
plt.grid(True)
plt.show()
```

```python
#finding the magnitude and phase response of system described by system function H(s).
import numpy as np
import matplotlib.pyplot as plt

def system_function(s, b_coeff, a_coeff):
    # Calculate H(s) as (b[0] + b[1]*s + b[2]*s^2) / (a[0] + a[1]*s + a[2]*s^2)
    numerator = sum(b * s**i for i, b in enumerate(b_coeff))
    denominator = sum(a * s**i for i, a in enumerate(a_coeff))
    return numerator / denominator

# Define system coefficients for H(s) = (b0 + b1 * s + b2 * s^2) / (a0 + a1 * s + a2 * s^2)
b_coeff = [1, 5, 6]  # Replace with your system's numerator coefficients
a_coeff = [1, 3, 2]  # Replace with your system's denominator coefficients

# Frequency range for analysis (omega values)
omega = np.linspace(0.1, 10, 500)
s = 1j * omega

# Compute H(jw)
H = np.array([system_function(si, b_coeff, a_coeff) for si in s])

# Magnitude and phase
magnitude = np.abs(H)
phase = np.angle(H)

# Plot magnitude response
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(omega, magnitude)
plt.xlabel('Frequency ()')
plt.ylabel('|H(j)|')
plt.title('Magnitude Response')
plt.grid(True)

# Plot phase response
plt.subplot(1, 2, 2)
plt.plot(omega, np.degrees(phase))  # Convert phase to degrees
plt.xlabel('Frequency ()')
plt.ylabel('Phase (degrees)')
plt.title('Phase Response')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sine(frequency, amplitude, phase, sample_rate, duration):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    sine_wave = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    return t, sine_wave

def generate_cosine(frequency, amplitude, phase, sample_rate, duration):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    cosine_wave = amplitude * np.cos(2 * np.pi * frequency * t + phase)
    return t, cosine_wave

# Parameters
frequency = 5         # Frequency in Hz
amplitude = 1         # Amplitude of the wave
phase = 0             # Phase shift in radians
sample_rate = 100     # Samples per second
duration = 1          # Duration in seconds

# Generate sine and cosine signals
t, sine_wave = generate_sine(frequency, amplitude, phase, sample_rate, duration)
_, cosine_wave = generate_cosine(frequency, amplitude, phase, sample_rate, duration)

# Plot the sine and cosine waves
plt.figure(figsize=(12, 5))

# Sine wave plot
plt.subplot(1, 2, 1)
plt.plot(t, sine_wave)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)

# Cosine wave plot
plt.subplot(1, 2, 2)
plt.plot(t, cosine_wave, color='orange')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_unit_step(length):
    """ Generate a Unit Step sequence """
    return np.where(np.arange(length) >= 0, 1, 0)

def generate_ramp(length):
    """ Generate a Ramp sequence """
    return np.maximum(0, np.arange(length))

def generate_exponential(length, amplitude, alpha):
    """ Generate an Exponential sequence """
    n = np.arange(length)
    return amplitude * np.exp(alpha * n)

# Parameters
length = 20               # Length of the sequence
amplitude = 1             # Amplitude for exponential
alpha = 0.1               # Growth rate for exponential

# Generate signals
unit_step = generate_unit_step(length)
ramp = generate_ramp(length)
exponential = generate_exponential(length, amplitude, alpha)

# Plot the signals
plt.figure(figsize=(15, 5))

# Unit Step plot
plt.subplot(1, 3, 1)
plt.stem(unit_step, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Unit Step Sequence')
plt.grid(True)

# Ramp plot
plt.subplot(1, 3, 2)
plt.stem(ramp, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="g-")
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Ramp Sequence')
plt.grid(True)

# Exponential plot
plt.subplot(1, 3, 3)
plt.stem(exponential, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Exponential Sequence')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def DFT(x):
    """Compute the Discrete Fourier Transform (DFT) of a sequence x."""
    N = len(x)
    X = [0] * N  # Initialize output array

    for k in range(N):
        real_sum = 0
        imag_sum = 0
        for n in range(N):
            angle = -2 * np.pi * k * n / N
            real_sum += x[n] * np.cos(angle)
            imag_sum += x[n] * np.sin(angle)
        X[k] = complex(real_sum, imag_sum)  # Combine real and imaginary parts

    return X

# Example input sequence
x = [1, 2, 3, 4]  # Replace with any input sequence

# Compute DFT
X = DFT(x)

# Compute magnitude and phase
magnitude = [abs(value) for value in X]
phase = [np.angle(value) for value in X]

# Plot the magnitude and phase
plt.figure(figsize=(12, 5))

# Magnitude plot
plt.subplot(1, 2, 1)
plt.stem(magnitude, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Magnitude')
plt.title('DFT Magnitude Spectrum')
plt.grid(True)

# Phase plot
plt.subplot(1, 2, 2)
plt.stem(phase, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Phase (radians)')
plt.title('DFT Phase Spectrum')
plt.grid(True)

plt.tight_layout()
plt.show()

# plot real and imaginary parts of DFT

plt.figure(figsize=(12, 5))

real_part = [value.real for value in X]
imag_part = [value.imag for value in X]

# Real part plot
plt.subplot(1, 2, 1)
plt.stem(real_part, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Real Part')
plt.title('DFT Real Part')
plt.grid(True)

# Imaginary part plot
plt.subplot(1, 2, 2)
plt.stem(imag_part, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Imaginary Part')
plt.title('DFT Imaginary Part')
plt.grid(True)

plt.tight_layout()
plt.show()

    
```

```python
import numpy as np
import matplotlib.pyplot as plt

def FFT(x):
    """Compute the Fast Fourier Transform (FFT) of a sequence x."""
    N = len(x)
    if N <= 1:
        return x  # Base case: if the input length is 1, return the input

    # Divide: Compute FFT of even and odd indexed parts
    even = FFT(x[0::2])
    odd = FFT(x[1::2])

    # Combine: Use symmetry property of twiddle factors
    T = [np.exp(-2j * np.pi * k / N) * odd[k] for k in range(N // 2)]
    return [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]

# Example input sequence (length should be a power of 2 for this recursive FFT)
x = [1, 2, 3, 4, 0, 0, 0, 0]  # Zero-padding to length 8 (next power of 2)

# Compute FFT
X = FFT(x)

# Compute magnitude and phase
magnitude = [abs(value) for value in X]
phase = [np.angle(value) for value in X]

# Plot the magnitude and phase
plt.figure(figsize=(12, 5))

# Magnitude plot
plt.subplot(1, 2, 1)
plt.stem(magnitude, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Magnitude')
plt.title('FFT Magnitude Spectrum')
plt.grid(True)

# Phase plot
plt.subplot(1, 2, 2)
plt.stem(phase, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Phase (radians)')
plt.title('FFT Phase Spectrum')
plt.grid(True)

plt.tight_layout()
plt.show()

# plot real and imaginary parts of FFT

plt.figure(figsize=(12, 5))

real_part = [value.real for value in X]
imag_part = [value.imag for value in X]


# Real part plot
plt.subplot(1, 2, 1)
plt.stem(real_part, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Real Part')
plt.title('DFT Real Part')
plt.grid(True)

# Imaginary part plot
plt.subplot(1, 2, 2)
plt.stem(imag_part, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel('Frequency bin (k)')
plt.ylabel('Imaginary Part')
plt.title('DFT Imaginary Part')
plt.grid(True)

plt.tight_layout()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the sine wave
frequency = 5  # Frequency of the sine wave in Hz
duration = 1.0  # Duration of the signal in seconds
sampling_freq_high = 50  # High sampling frequency in Hz (above Nyquist rate)
sampling_freq_low = 8    # Low sampling frequency in Hz (below Nyquist rate)

# Time axis for the continuous sine wave
t_continuous = np.linspace(0, duration, 1000)
# Generate the continuous sine wave
sine_wave_continuous = np.sin(2 * np.pi * frequency * t_continuous)

# Time axis for high sampling rate (above Nyquist frequency)
t_high = np.arange(0, duration, 1 / sampling_freq_high)
# Sampled sine wave with high sampling rate
sine_wave_high = np.sin(2 * np.pi * frequency * t_high)

# Time axis for low sampling rate (below Nyquist frequency, causes aliasing)
t_low = np.arange(0, duration, 1 / sampling_freq_low)
# Sampled sine wave with low sampling rate (aliasing effect)
sine_wave_low = np.sin(2 * np.pi * frequency * t_low)

# Plotting
plt.figure(figsize=(12, 8))

# Plot continuous sine wave
plt.subplot(3, 1, 1)
plt.plot(t_continuous, sine_wave_continuous, label="Continuous Sine Wave")
plt.title("Continuous Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()

# Plot sine wave sampled above Nyquist rate
plt.subplot(3, 1, 2)
plt.plot(t_continuous, sine_wave_continuous, label="Continuous Sine Wave", color="lightgray")
plt.stem(t_high, sine_wave_high, linefmt="b-", markerfmt="bo", basefmt=" ", use_line_collection=True)
plt.title(f"Sine Wave Sampled at {sampling_freq_high} Hz (Above Nyquist Rate)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()

# Plot sine wave sampled below Nyquist rate
plt.subplot(3, 1, 3)
plt.plot(t_continuous, sine_wave_continuous, label="Continuous Sine Wave", color="lightgray")
plt.stem(t_low, sine_wave_low, linefmt="r-", markerfmt="ro", basefmt=" ", use_line_collection=True)
plt.title(f"Sine Wave Sampled at {sampling_freq_low} Hz (Below Nyquist Rate - Aliasing)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()
```
