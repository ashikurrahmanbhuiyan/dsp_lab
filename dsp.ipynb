```python
import matplotlib.pyplot as plt

def convolution(x, h):
    N = len(x)
    M = len(h)
    output_size = N + M - 1
    y = [0] * output_size

    for n in range(output_size):
        for k in range(N):
            if 0 <= n - k < M:
                y[n] += x[k] * h[n - k]
    return y

# Define sequences
x = [1, 2, 3]  # Input sequence
h = [4, 5, 6]  # Impulse response

# Get convolution result
result = convolution(x, h)

# Plot the result
plt.figure(figsize=(10, 4))
plt.stem(result, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="b-")
plt.xlabel("Index (n)")
plt.ylabel("Amplitude")
plt.title("Convolution of x and h")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def autocorrelation(x):
    N = len(x)
    autocorr = [0] * N
    for m in range(N):
        autocorr[m] = sum(x[n] * x[n + m] for n in range(N - m))
    return autocorr

# Define sequence
x = [1, 2, 3, 4]  # Example sequence

# Compute autocorrelation
result = autocorrelation(x)

# Plot the result
plt.figure(figsize=(10, 4))
plt.stem(result, use_line_collection=True, basefmt=" ", markerfmt="o", linefmt="r-")
plt.xlabel("Lag (m)")
plt.ylabel("Autocorrelation")
plt.title("Autocorrelation of the Sequence")
plt.grid(True)
plt.show()
```
